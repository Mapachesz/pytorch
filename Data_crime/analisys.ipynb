## **Graphs**


```python
#Replace null values with the mean of the column
crime_df.fillna(crime_df.select_dtypes(include=np.number).mean(), inplace=True)
```

```python
#Obtain number of numeric columns
len(crime_df.select_dtypes(np.number).columns)
```

*Distribution of numeric values*

```python

fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(20, 20))
axes = axes.flat
numeric_columns = crime_df.select_dtypes(include=np.number).columns

for i, colum in enumerate(numeric_columns):
    sns.histplot(
        data     = crime_df,
        x        = colum,
        stat     = "count",
        kde      = True,
        color    = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        line_kws = {'linewidth': 2},
        alpha    = 0.3,
        ax       = axes[i]
    )
    axes[i].set_title(colum, fontsize = 7, fontweight = "bold")
    axes[i].tick_params(labelsize = 6)
    axes[i].set_xlabel("")
    
    
fig.tight_layout()
plt.subplots_adjust(top = 0.9)
fig.suptitle('Distribution of numeric values', fontsize = 10, fontweight = "bold")
```

*Correlation matrix*

```python
numeric_columns = crime_df.select_dtypes(include=np.number).columns

matriz_corr = pd.DataFrame(numeric_columns).corr('pearson')

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))

sns.heatmap(
    matriz_corr,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 6},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 8)
```
