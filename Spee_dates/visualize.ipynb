```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("speed_data.csv")
data = data.dropna()
```

```python
data.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
atributos = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar', 'like']

# Establecer el tamaño del gráfico
plt.figure(figsize=(15, 10))

# Crear un gráfico para cada atributo
for i, atributo in enumerate(atributos, 1):
    plt.subplot(3, 3, i)  # Ajusta el número de filas y columnas según la cantidad de atributos
    sns.scatterplot(x=atributo, y='prob', data=data)
    plt.title(f'Probabilidad de Match vs {atributo}')
    plt.xlabel(f'{atributo}')
    plt.ylabel('Probabilidad de Match')

plt.tight_layout()
plt.show()
```

```python
#Grafico con solo datos numericos

# Seleccionar solo columnas numéricas para la correlación
datos_numericos = data.select_dtypes(include=[np.number])

# Calcular la matriz de correlación solo con datos numéricos
correlacion = datos_numericos.corr()


# Creando el mapa de calor con Seaborn
plt.figure(figsize=(12, 10))  # Ajusta el tamaño del gráfico según necesites
sns.heatmap(correlacion, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Mapa de Calor de Correlación entre Características")
plt.show()
```
